```python
# 9
# Problem 9: Determine if a linked list is a palindrome.
# Input: 1 -> 2 -> 2 -> 1
# Output: True

class Node :
    def __init__(self, data = None, next = None):
        self.data = data
        self.next = next
    def set_data(self, data):
        self.data = data
    def get_data(self):
        return self.data
    def set_next(self, next= None):
        self.next = next
    def get_next(self):
        return self.next

def traverse(head):
    temp = head
    while(temp != None):
        print(temp.get_data(),end = "")
        if temp.get_next() != None:
            print("->", end="")
        else:
            print()
        temp = temp.get_next()

def length(head):
    leng = 0
    temp = head
    while(temp != None):
        leng += 1
        temp = temp.get_next()
    return leng

def index_node(head, n):
    i = 0
    temp = head
    while i < n:
        i += 1
        temp = temp.get_next()
    return temp

def is_pallidrome_linked_list(head):
    is_pallindrome = True
    n = length(head)
    for i in range(n):
        if index_node(head, i).get_data() != index_node(head, n - i - 1).get_data():
            is_pallindrome = False
    return is_pallindrome

head = Node(1)
node2 = Node(2)
node3 = Node(3)
node4 = Node(3)
node5 = Node(2)
node6 = Node(1)

head.set_next(node2)
node2.set_next(node3)
node3.set_next(node4)
node4.set_next(node5)
node5.set_next(node6)

traverse(head)
print(is_pallidrome_linked_list(head))
```

```python
# 16
# Problem 16: Search for a target value in a rotated sorted array.
# Input: nums = [4, 5, 6, 7, 0, 1, 2], target = 0
# Output: 4

def search_in_rotated_sorted_arr(arr, target):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        if arr[low] <= arr[mid]:
            if arr[low] <= target < arr[mid]:
                high = mid - 1
            else:
                low = mid + 1
        else:
            if arr[mid] < target <= arr[high]:
                low = mid + 1
            else:
                high = mid - 1
    return -1

array = [4, 5, 6, 7, 0, 1, 2]
target = 0
print(search_in_rotated_sorted_arr(array, target))
```

```python
# 17
# Problem 17: Find the peak element in an array. A peak element is greater than its neighbors.
# Input: nums = [1, 2, 3, 1]
# Output: 2 (index of peak element)

def find_peak_elements(nums):
    peak_elements = []
    for i in range(len(nums)):
        if i == 0:
            if nums[i] > nums[i + 1]:
                peak_elements.append(i)
        elif i == len(nums) - 1:
            if nums[i] > nums[i - 1]:
                peak_elements.append(i)
        else:
            if nums[i] > nums[i - 1] and nums[i] > nums[i + 1]:
                peak_elements.append(i)
    return peak_elements

nums = [1, 2, 3, 1]
print(find_peak_elements(nums))
```

```python
# 18
# Problem 18: Given a m x n matrix where each row and column is sorted in ascending
# order, count the number of negative numbers.
# matrix = [[-4, -3, -1, 1], [-2, -2,  1, 2], [-1,  1,  2, 3], [ 1,  2,  4, 5]]
# Output: 8

def count_negative(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    row = 0
    col = cols - 1
    count = 0
    while row < len(matrix) and col >=0:
        if matrix[row][col] < 0:
            count += col + 1
            row += 1
            col = cols -1
        else:
            col -= 1
    return count

matrix = [
    [-4, -3, -1, 1],
    [-2, -2,  1, 2],
    [-1,  -1, -2, 3],
    [ 1,  2,  4, 5]
]
print(count_negative(matrix))
```

```python
# 19
# Problem 19: Given a 2D matrix sorted in ascending order in each row, and the first integer of each row is
# greater than the last integer of the previous row, determine if a target value is present in the matrix.
# Input: matrix = [[1, 3, 5, 7], [10, 11, 16, 20], [23, 30, 34, 60]], target = 3
# Output: True

def binary_search_in_sorted_matrix(matrix, target):
    row_length = len(matrix[0]) - 1
    def helper_row_search_matrix():
        low = 0
        high = len(matrix) - 1
        while low <= high:
            mid = (low + high ) // 2
            if(target < matrix[mid][0]):
                high = mid - 1
            elif(target > matrix[mid][row_length]):
                low = mid + 1
            else:
                return mid
    
    target_row = helper_row_search_matrix()
    low = 0
    high = row_length
    while low != high:
        mid = (low + high) // 2
        if(target > matrix[target_row][mid]):
            low = mid + 1
        elif(target < matrix[target_row][mid]):
            high = mid - 1
        else:
            return target_row,mid

matrix = [[1, 3, 5, 7], [10, 11, 16, 20], [23, 30, 34, 60]]
target = 3
print(binary_search_in_sorted_matrix(matrix,3))
```

```python
# 20
# Problem 20: Find Median in Two Sorted Arrays
# Given two sorted arrays, find the median of the combined sorted array.
# Input: nums1 = [1, 3], nums2 = [2]
# Output: 2.0

def merge_sorted_lists(list1, list2):
    merged_list = []
    i = j = 0
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1
    while i < len(list1):
        merged_list.append(list1[i])
        i += 1
    while j < len(list2):
        merged_list.append(list2[j])
        j += 1
    return merged_list

def median_in_merge_sorted_arrays(arr1, arr2):
    combined = merge_sorted_lists(arr1, arr2)
    combined_length = len(combined)
    if combined_length % 2 == 0:
        return (combined[int(combined_length / 2)] + combined[int(combined_length / 2 - 1)]) / 2
    else:
        return combined[int((combined_length - 1) / 2)]


list1 = [1, 3, 5, 7]
list2 = [2, 4, 6,8]
merged_list = merge_sorted_lists(list1, list2)
print("Merged List:", merged_list)
print(median_in_merge_sorted_arrays(list1, list2))
```

```python
# 21
# Problem 21: Given a sorted character array and a target letter, find the smallest letter in the array that is
# greater than the target.
# Input: letters = ['c', 'f', 'j'], target = a
# Output: 'c'

def sorted_char_just_greater(arr,target):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if (arr[mid] > target):
            high = mid - 1
            ans = mid
        else:
            low = mid + 1
    if (arr[ans] < target):
        return arr[0]
    else:
        return arr[ans]

letters = ['c', 'f', 'j']
target = 'g'
print(sorted_char_just_greater(letters, target))
```

```python
# 22
# Problem 22: Given an array with n objects colored red, white, or blue, sort them in-place so that objects of
# the same color are adjacent, with the colors in the order red, white, and blue.
# Input: nums = [2, 0, 2, 1, 1, 0]
# Output: [0, 0, 1, 1, 2, 2]

def sort012(arr):
    low = 0
    mid = 0
    high = len(arr)-1

    while mid<high:
        if arr[mid] == 0:
            arr[low], arr[mid] = arr[mid], arr[low]
            low += 1
            mid += 1
        elif arr[mid] == 1:
            mid += 1
        else:
            arr[mid], arr[high] = arr[high], arr[mid]
            high -= 1
    return arr

print(sort012([0,2,0,1,1,2,0,1,0,1,2]))
```

```python
# 23
# Problem 23: Find the kth largest element in an unsorted array.
# Input: nums = [3, 2, 1, 5, 6, 4], k = 2
# Output: 5

def find_kth_largest(arr, k):
    arr.sort()
    return arr[-k]

arr = [3, 1, 4, 5, 2]
k = 2
if k == 1:
    print(f"{k}st largest element:", find_kth_largest(arr, k))
elif k == 2:
    print(f"{k}nd largest element:", find_kth_largest(arr, k))
else:
    print(f"{k}st largest element:", find_kth_largest(arr, k))
```

```python
# 25
# Problem 25: Given an array of integers, calculate the sum of all its elements.
# Input: [1, 2, 3, 4, 5]
# Output: 15

def calculate_sum(arr):
    total_sum = 0
    for num in arr:
        total_sum += num
    return total_sum

arr = [1, 2, 3, 4, 5]
total = calculate_sum(arr)
print("Sum of all elements:", total)
```

```python
# 26
# Problem 26: Find the maximum element in an array of integers.
# Input: [5, 3, 8, 2, 7, 4], target = 8
# Output: 2

def find_maximum(arr):
    if not arr:
        return None
    max_element = arr[0]
    for num in arr:
        if num > max_element:
            max_element = num
    return max_element

arr = [3, 7, 2, 9, 5]
maximum = find_maximum(arr)
print("Maximum element:", maximum)
```

```python
# 27
# Problem 27: Implement linear search to find the index of a target element in an array.
# Input: [5, 3, 8, 2, 7, 4], target = 8
# Output: 2

def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

arr = [3, 6, 2, 8, 4]
target = 8
index = linear_search(arr, target)
if index != -1:
    print("Target", target, "found at index:", index)
else:
    print("Target", target, "not found in the array.")
```

```python
# 28
# Problem 28 Calculate the factorial of a given number.
# Input: 5
# Output: 120 (as 5! = 5 * 4 * 3 * 2 * 1 = 120)

def factorial(num):
    fact = 1
    for i in range(fact, num + 1):
        fact *= i
    return fact
print(factorial(5))
```

```python
# 29
# Problem 29: Check if a given number is a prime number.
# Input: 7
# Output: True

def is_prime(n):
    if n <= 1:
        return False
    elif n <= 3:
        return True
    elif n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

# Example usage:
num = 29
if is_prime(num):
    print(num, "is a prime number.")
else:
    print(num, "is not a prime number.")
```

```python
# 30
# Problem 30: Generate the Fibonacci series up to a given number n.
# Input: 8
# Output: [0, 1, 1, 2, 3, 5, 8, 13]

def generate_fibonacci_series(n):
    fibonacci_series = [0, 1]
    while fibonacci_series[-1] + fibonacci_series[-2] <= n:
        next_number = fibonacci_series[-1] + fibonacci_series[-2]
        fibonacci_series.append(next_number)
    return fibonacci_series

n = 100
fibonacci_series = generate_fibonacci_series(n)
print("Fibonacci series up to", n, ":", fibonacci_series)
```

```python
# 31
# Problem 31: Calculate the power of a number using recursion.
# Input: base = 3, exponent = 4
# Output: 81 (as 3^4 = 3 * 3 * 3 * 3 = 81)
def power(num, exp):
    if exp == 1:
        return num
    elif exp % 2 == 0:
        return power(num, exp / 2) * power(num, exp / 2)
    else:
        return num * power(num, exp // 2) * power(num, exp // 2)

print(power(2, 10))
```

```python
# 32
# Problem 32: Reverse a given string.
# Input: "hello"
# Output: "olleh"

def reverse_string(s):
    reversed_str = ""
    for i in range(len(s) - 1, -1, -1):
        reversed_str += s[i]
    return reversed_str

s = "hello"
reversed_s = reverse_string(s)
print("Original:", s)
print("Reversed:", reversed_s)
```
